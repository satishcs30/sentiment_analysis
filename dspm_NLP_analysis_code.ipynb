### Text Clean up 
1. We carry out the sequential steps of cleaning the data before it can be tokenized for the sentiment analysis. 
2. A clean-up stage is essential because the raw tweets are susceptible to nonuniform language patterns which could be miss classified by our analyzer. 

```python
# Pefroming String Operations on the Tweets data
# Following operations will be perfromed
#1. Lower the case
#2. Expanding the contractions
#3. Regex based data cleanup
    #3a Remove URLs
    #3b Remove Username 
    #3c Remove Hashtags 
    #3d Remove White spaces
    #3e Remove Digits
    #3f Removing puncuations
    
#4a. Removing the stop wrods
#4b. Removing the frequenty used words.

import re
import string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from string import digits


#1. Make lower case 
# This metod is added so that the reader can visualize the steps
def toLowerCase(text):
    return text.lower()

#2 Exapanding contractions 
contractions_dict = { "ain't": "are not","'s":" is","aren't": "are not","can't": "cannot","can't've": "cannot have","'cause": "because","could've": "could have","couldn't": "could not","couldn't've": "could not have", "didn't": "did not","doesn't": "does not","don't": "do not","hadn't": "had not","hadn't've": "had not have","hasn't": "has not","haven't": "have not","he'd": "he would","he'd've": "he would have","he'll": "he will", "he'll've": "he will have","how'd": "how did","how'd'y": "how do you","how'll": "how will","I'd": "I would", "I'd've": "I would have","I'll": "I will","I'll've": "I will have","I'm": "I am","I've": "I have", "isn't": "is not","it'd": "it would","it'd've": "it would have","it'll": "it will","it'll've": "it will have", "let's": "let us","ma'am": "madam","mayn't": "may not","might've": "might have","mightn't": "might not", "mightn't've": "might not have","must've": "must have","mustn't": "must not","mustn't've": "must not have", "needn't": "need not","needn't've": "need not have","o'clock": "of the clock","oughtn't": "ought not","oughtn't've": "ought not have","shan't": "shall not","sha'n't": "shall not","shan't've": "shall not have","she'd": "she would","she'd've": "she would have","she'll": "she will", "she'll've": "she will have","should've": "should have","shouldn't": "should not", "shouldn't've": "should not have","so've": "so have","that'd": "that would","that'd've": "that would have", "there'd": "there would","there'd've": "there would have", "they'd": "they would","they'd've": "they would have","they'll": "they will","they'll've": "they will have", "they're": "they are","they've": "they have","to've": "to have","wasn't": "was not","we'd": "we would","we'd've": "we would have","we'll": "we will","we'll've": "we will have","we're": "we are","we've": "we have", "weren't": "were not","what'll": "what will","what'll've": "what will have","what're": "what are", "what've": "what have","when've": "when have","where'd": "where did", "where've": "where have","who'll": "who will","who'll've": "who will have","who've": "who have","why've": "why have","will've": "will have","won't": "will not","won't've": "will not have", "would've": "would have","wouldn't": "would not","wouldn't've": "would not have","y'all": "you all", "y'all'd": "you all would","y'all'd've": "you all would have","y'all're": "you all are","y'all've": "you all have", "you'd": "you would","you'd've": "you would have","you'll": "you will","you'll've": "you will have", "you're": "you are","you've": "you have"}
# Regular expression for finding contractions
contractions_re = re.compile('(%s)' % '|'.join(contractions_dict.keys()))

def expand_contractions(text,contractions_dict=contractions_dict):
    def replace(match):
        return contractions_dict[match.group(0)]
    return contractions_re.sub(replace, text)


#3 Perfroming multiple regex based clean up of data 

#3a Removing the URL Links with a regular expresion.
def removing_URLs(text):
    return re.sub(r'((www\.[^\s]+)|(http?://[^\s]+)|(https?://[^\s]+))', ' ', text)
#3b Removing tweets that involve the use of the at (@)    
def removing_username(text):
    return re.sub("@[\w]*", ' ', text)

#3c Replacing the #Links with a regular expresion.    
def removing_hashtags(text):
    return re.sub(r'#([^\s]+)', r'\1', text)
    
#3d Removing white spaces
def removing_whitespaces(text):
    return re.sub('[\s]+', ' ', text)

#3e Removing digits
def removing_digits(text):
    remove_digits = str.maketrans('', '', digits)
    res = text.translate(remove_digits)
    return res
    
#3f Removing special characters and puncuations
def removing_punctuations_specialcharacters(text):
    return re.sub('[^A-Za-z0-9]+', ' ', text)

#4a. Removing the stop words

def remove_stopWords(text):
    remove_punc = [word for word in text.split() if word not in stopwords.words('english')]
    remove_punc = ' '.join(remove_punc)
    return remove_punc

#4b Removing frequently used ans swear words 
frequent_words = ['s8','galaxy','samsung','com','twitter','pic','dlvr','tt','ift','phone','new', 'iphone', '8', 'com', 'apple', 'twitter', 'pic', 'dlvr', 'new', 'tt', 'ift', 'iphone','x','8','youtu','video','apple','com','pocketnow','samsung', 'fb', '2017', 'via', 'goo', 'gl', 'vs', 'news', '4wtdvs65c9g', 'iphonex', 'iphone x', 'ly', 'google', 'fuck']
def remove_frequentWords(text):
    remove_frequent = [word for word in text.split() if word not in frequent_words]
    remove_frequent = ' '.join(remove_frequent)
    return remove_frequent

    
```

### Calling all above cleanup operations in sequence for a given text

```python
#. Calling all above operations in sequence for a given sentiment
def mycustom_textcleanup(text):
    temp = text
    #Lower case 
    temp = toLowerCase(temp)
    
    #Expand Contractions 
    temp = expand_contractions(temp)
    
    #Perfrom regex based replacements 
    temp = removing_URLs(temp)
    temp = removing_username(temp)
    temp = removing_hashtags(temp)
    temp = removing_whitespaces(temp)
    temp = removing_digits(temp)
    temp = removing_punctuations_specialcharacters(temp)
   
    #Remove Stop words and frequently used words 
    temp = remove_stopWords(temp)
    temp = remove_frequentWords(temp)
    return temp
```

### What were the most important attributes for each product?

```python
#Identifying top attributes for each phone -> nouns, adjectives and overall words
product_dict = {'s8': "Samsung S8", 'i8': "Iphone 8", 'ix': "Iphone X"} 
overall = {'s8':[], 'i8' : [], 'ix' : []} ## product 1: [], 2:[]
nouns = {'s8':[], 'i8' : [], 'ix' : []}
adjectives = {'s8':[], 'i8' : [], 'ix' : []}

def important_attributes(df):
    for i in product_dict.keys(): # iterate through the product labels
        list_of_tokenizedtweets = list(df[df['tweet_label'] == i]['tokenized tweets']) # i = 1 ==> [s8, galaxy, good, s8 galaxy bad]]
        for tweet_words_list  in list_of_tokenizedtweets: #[s8 galaxy good]
            tokenized_tagged = nltk.pos_tag(tweet_words_list) # [(s8, NN), (galaxy, noun), (good, adj)]
            for w in tokenized_tagged:
                if w[1][0] in ["J", "JJ"]: # Iddentify the Adjective Tag 
                    # read the elemnts in the dict and append 
                    adjectives[i].append(w[0]) 
                if w[1][0] in ["N", "NN"]: # Identify the Noun Tag
                    nouns[i].append(w[0])
                overall[i].append(w[0])
```

```python
#defing textBlob function
from textblob import TextBlob
def get_tweet_sentiment(text):
    
        analysis = TextBlob(text).sentiment
        
        # set sentiment
        if analysis.polarity > 0:
            senti_rate = 1
        elif analysis.polarity == 0:
            senti_rate = 0
        else:
            senti_rate = -1
        return senti_rate, analysis.polarity
```

```python
sentiment_text_blob = []
polarity_text_blob = []

for i in combined_df["cleaned Sound Bite Text"]:
    senti_rate, polarity = get_tweet_sentiment(i)
    sentiment_text_blob.append(senti_rate)
    polarity_text_blob.append(polarity)
    
#Add the sentiment and polarity from textblob to dataframe
combined_df['sentiment_text_blob'] = sentiment_text_blob
combined_df['polarity_text_blob'] = polarity_text_blob
```

### Evaluating Sentiment and polarity from VADER

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
def vader_compound_score(c):
    if c>0:
        return 1
    elif c==0:
        return 0
    else:
        return -1
    return 0
```

```python
combined_df['scores'] = combined_df['cleaned Sound Bite Text'].apply(lambda review: sid.polarity_scores(review))
combined_df['compound']  = combined_df['scores'].apply(lambda score_dict: score_dict['compound'])
combined_df['comp_score'] = combined_df['compound'].apply(vader_compound_score)
combined_df.head()
```

```python
#plot text_blob polarity results
combined_df.pivot(columns='tweet_label').polarity_text_blob.plot(kind = 'hist', alpha=0.5)
plt.title('Text Blob sentiment polarity')
plt.xlabel('polarity')
plt.show()
```

```python
#Plot Vader polarity results 
combined_df.pivot(columns='tweet_label').compound.plot(kind = 'hist', alpha = 0.5)
plt.title('VADER sentiment polarity')
plt.xlabel('polarity')
plt.ylim((0, 8000))
plt.show()
```

```python
#Additional non-zero polarity datapoints captured by VADER
txtBlb_non_zero = combined_df[combined_df.polarity_text_blob !=0]["Sound Bite Text"].count()
vader_non_zero = combined_df[combined_df.compound !=0]["Sound Bite Text"].count()
overall_tweets_count = combined_df["Sound Bite Text"].count()

print("% of Text Blob non-zero tweets: ", txtBlb_non_zero/overall_tweets_count*100)
print("% of VADER non-zero tweets: ", vader_non_zero/overall_tweets_count*100)
print("% of additional non-zero tweets captured by Vader: ", (vader_non_zero - txtBlb_non_zero)/overall_tweets_count*100)
```

```python
###Functions for positive and negative sentiment categorization
adjectives_pos = {'s8':[], 'i8' : [], 'ix' : []}
nouns_pos = {'s8':[], 'i8' : [], 'ix' : []}
adjectives_neg = {'s8':[], 'i8' : [], 'ix' : []}
nouns_neg = {'s8':[], 'i8' : [], 'ix' : []}

def positive_sentiment_analyzer(df):
    for i in product_dict.keys(): # iterate through the product labels
        list_of_tokenizedtweets = list(df[(df['comp_score'] > 0)  & (df['tweet_label'] == i)]['tokenized tweets']) 
        for tweet_words_list  in list_of_tokenizedtweets: #[s8 galaxy good]
            tokenized_tagged = nltk.pos_tag(tweet_words_list) # [(s8, NN), (galaxy, noun), (good, adj)]
            for w in tokenized_tagged:
                if w[1][0] in ["J", "JJ"]: # Identify the Adjective Tag 
                    # read the elemnts in the dict and append 
                    adjectives_pos[i].append(w[0]) 
                if w[1][0] in ["N", "NN"]: # Identify the Noun Tag
                    nouns_pos[i].append(w[0])

def negative_sentiment_analyzer(df):
    for i in product_dict.keys(): # iterate through the product labels
        list_of_tokenizedtweets = list(df[(df['comp_score'] < 0)  & (df['tweet_label'] == i)]['tokenized tweets']) 
        for tweet_words_list  in list_of_tokenizedtweets: #[s8 galaxy good]
            tokenized_tagged = nltk.pos_tag(tweet_words_list) # [(s8, NN), (galaxy, noun), (good, adj)]
            for w in tokenized_tagged:
                if w[1][0] in ["J", "JJ"]: # Identify the Adjective Tag 
                    # read the elemnts in the dict and append 
                    adjectives_neg[i].append(w[0]) 
                if w[1][0] in ["N", "NN"]: # Identify the Noun Tag
                    nouns_neg[i].append(w[0])
                    
                    
                    
positive_sentiment_analyzer(combined_df)
negative_sentiment_analyzer(combined_df)
```

### Part 3 How did consumers feel about the Quality, Price and “Value” for each



```python
## Feature Profilers code: 
## This code identifies the attributes and tags them for quality, price and Value

lookup_dictionary = {
    'quality':["design", "cover", "screen", "battery", "look","tough", "clear", "active", "slow","camera", "good", "bad", "poor", "fast","display", "quality"], 
    'price':["cheap", "expensive", "cost", "price", "pricey"],
    'value':["value", "worth", "cost", "evaluation", "merit", "value"]
                    }
def identify_attributes(text, feature ):
    out = False 
    lookup = lookup_dictionary[feature]
    return any(word in lookup for word in text)
```

```python
combined_df["Quality"] = combined_df['tokenized tweets'].apply(lambda text: identify_attributes(text, 'quality'))
combined_df["Price"] = combined_df['tokenized tweets'].apply(lambda text: identify_attributes(text, 'price'))
combined_df["Value"] = combined_df['tokenized tweets'].apply(lambda text: identify_attributes(text, 'value'))
```

```python
###

def plot_attributes_histograms(dataframe):  
    for i in product_dict.keys():
        df_quality = dataframe[(dataframe["Quality"] == True) & (dataframe["tweet_label"]== i)]["comp_score"]
        plt.title(product_dict[i] + " Quality Sentiment Rating ")
        plt.hist(df_quality, bins=3)
        plt.show() 
        df_price = dataframe[(dataframe["Price"] == True) & (dataframe["tweet_label"]== i)]["comp_score"]
        plt.title(product_dict[i] + " Price Sentiment Rating")
        plt.hist(df_price, bins=3)
        plt.show() 
        df_value = dataframe[(dataframe["Value"] == True) & (dataframe["tweet_label"]== i)]["comp_score"]
        plt.title(product_dict[i] + " Value Sentiment Rating")
        plt.hist(df_value, bins=3)
        plt.show()
```

```python
plot_attributes_histograms(combined_df)
```

**Samsung Summary:**
* Distributions for quality and value were unsurprising because the general brand perception for Samsung is that their phones are dependable and consistent
* It is a well known fact that the pricepoint is lower than Apple products, which is often a point of argument in the big Apple vs Android debate. 
* The slight positive skew in the price sentiment is likely because of that same fact, but it didn’t have much buzz because it’s an inevitability at this point. 
   

**iPhone 8 Summary:**
* Strong positive sentiment for both quality and value, but mostly neutral with a slightly negative sentiment for the price
* It’s a well known fact that iPhones are on the more expensive side of smartphones, so it makes sense that again there wasn’t much buzz complaining about that to create a strong negative sentiment about price


**iPhone X Summary:**
* The strong positive sentiment about the price is surprising, because it was by far the most expensive iPhone ever released at that point, with even higher price points internationally 
* Both the iPhone x and iPhone 8 had strong positive sentiment for quality and value, which makes sense due to the very strong customer base for the Apple ecosystem.
